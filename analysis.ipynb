# Data Prepocessing

## TONO Dataset Quality Assessment

Output of the quality assesment on the TONO dataset

```python
import pandas as pd

df = pd.read_csv('src/data/QFIQ_assessment/tono_assessment.csv', delimiter=';')
# Remove the ../../src/data/TONO/release/ prefix from the Filename column rows
df['Filename'] = df['Filename'].str.replace(r'../../src/data/TONO/release/', '', regex=True)
df['ContrastElement'] = df['Filename'].str.split('/').str[0]
df['Filename'] = df['Filename'].str.replace(r'^[^/]+/', '', regex=True)
df
```

## ONOT Dataset Quality Assessment

21 images were assessed from the ONOT dataset as they are ICAO-compliant and therefore can provide the LLM with the larger spectrum of references to what is acceptable

```python
# Show the output of the ONOT assessment
df_onot = pd.read_csv('src/data/QFIQ_assessment/onot_assessment.csv', delimiter=';')
# Remove the ../../src/data/TONO/release/ prefix from the Filename column rows
df_onot['Filename'] = df_onot['Filename'].str.replace(r'../../src/data/ONOT/S115379105/', '', regex=True)
df_onot['ContrastElement'] = None

df_onot
```

Merging TONO and ONOT assessments

```python
# Merge the two dataframes
df_combined = pd.concat([df, df_onot], ignore_index=True)
# Move the ContrastElement as a second column
df_combined = df_combined[['Filename', 'ContrastElement'] + [col for col in df_combined.columns if col not in ['Filename', 'ContrastElement']]]
df_combined
```

Keep only the columns that output the scalar values, file names, and contrast elements

```python

cols_to_keep = ['Filename', 'ContrastElement'] + [col for col in df_combined.columns if col.endswith('scalar')]
df_combined = df_combined[cols_to_keep]
df_combined
```

For each of the contrast element group, 6 images will be chosen with the variety of the utility score for providing the testing base for the LLM. Additionally 6 images from the ONOT database were chosen.

```python
# In each of the contrast elements group, get 3 images with the highest UnifiedQualityScore.scalar, 3 with the lowest UnifiedQualityScore.scalar, and 3 with the median UnifiedQualityScore.scalar. or closest to it.
def get_top_bottom_median(group):
    top = group.nlargest(2, 'UnifiedQualityScore.scalar')
    bottom = group.nsmallest(2, 'UnifiedQualityScore.scalar')
    median_value = group['UnifiedQualityScore.scalar'].median()
    median = group.iloc[(group['UnifiedQualityScore.scalar'] - median_value).abs().argsort()[:2]]
    
    return pd.concat([top, bottom, median])

# Apply the function to each group of ContrastElement
df_in_context_train = df_combined.groupby('ContrastElement').apply(get_top_bottom_median).reset_index(drop=True)

# Add 6 top rows from the ONOT assesnment: where ContrastElement is None
df_in_context_train = pd.concat([df_in_context_train, df_combined[df_combined['ContrastElement'].isnull()].head(6)], ignore_index=True)

# Save the resulting DataFrame to a CSV file
df_in_context_train.to_csv('src/data/QFIQ_assessment/tono_assessment_contrast_element_test.csv', index=False)
```

# In-context learning

### Train data preparation

The goal is to provide the descriptions for the train data, so that they can be used for prompting LLM for in-context learning.

```python
# Read the saved CSV file
df_in_context_train = pd.read_csv('src/data/QFIQ_assessment/tono_assessment_contrast_element_test.csv')
df_in_context_train
```

Combine all the scalar values from `in_context_train` dataset into one column `OFIQResults`

```python
df_in_context_train['OFIQResults'] = df_in_context_train.apply(
    lambda row: {col: row[col] for col in df_in_context_train.columns if col.endswith('scalar')}, axis=1
)
# Removing scalar columns
df_in_context_train = df_in_context_train.drop(columns=[col for col in df_in_context_train.columns if col.endswith('scalar')])
```

Provide the descriptions based on the contrast element using the function `generate_description`

```python
from src.scripts.description import generate_description


df_in_context_train['Description'] = df_in_context_train.apply(
    lambda row: generate_description(row['ContrastElement'], row['OFIQResults']), axis=1
)

show_columns = ['Filename', 'ContrastElement', 'Description']
pd.set_option('display.max_colwidth', None)
df_in_context_train[show_columns]
```

NOTE: In order to provide the valid description, the correlation diagram was created, to analyze which scalar values have the highest impact on different `ContrastElements`. 

```python
# Create a heatmap as a correlation matrix for each of the contrast elemnt which scalar values affect it the most in the df_combined

import seaborn as sns
import matplotlib.pyplot as plt 

df_combined_corr= df_combined.copy()
df_combined_corr = df_combined_corr.drop(columns=['Filename'])

# Do one-hot encoding for the ContrastElement column
df_combined_corr = pd.get_dummies(df_combined_corr, columns=['ContrastElement'])
correlation_matrix = df_combined_corr.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    square=True,
    cbar_kws={"shrink": .8},
    annot_kws={"size": 6}  # Smaller font size for numbers
)
plt.title('Correlation Matrix of Scalar Values')
plt.xticks(rotation=60, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
```

### Test data preparation 

Choose a testing dataset from the `df_combined`. For testing purposes 2 images from each of the `ContrastElement` are being chosen in addition to 2 complaint images.  

```python
df_in_context_test = df_combined.copy()

# Removing images that are already in the training set, so we have a separate test set
df_in_context_test = df_in_context_test[~df_in_context_test['Filename'].isin(df_in_context_train['Filename'])]

# Grouping by ContrastElement and taking 2 images from each group
df_in_context_test = df_in_context_test.groupby('ContrastElement').head(2)

# Adding 2 images from ONOT assessment rows
df_in_context_test = pd.concat([df_in_context_test, df_combined[df_combined['ContrastElement'].isnull()].tail(2)], ignore_index=True)

df_in_context_test
```

Combine all the scalar values from  `in_context_test` dataset into `OFIQResults`, so that it matches `in_context_train` dataset

```python
df_in_context_test['OFIQResults'] = df_in_context_test.apply(
    lambda row: {col: row[col] for col in df_in_context_test.columns if col.endswith('scalar')}, axis=1
)
# Removing scalar columns
df_in_context_test = df_in_context_test.drop(columns=[col for col in df_in_context_test.columns if col.endswith('scalar')])
```

Save both train and test datasets into the JSONL files

```python

df_in_context_train.to_json('src/data/QFIQ_assessment/in_context_train_data.jsonl', orient='records', lines=True, force_ascii=False)
df_in_context_test.to_json('src/data/QFIQ_assessment/in_context_test_data.jsonl', orient='records', lines=True, force_ascii=False)
```
