# Data Prepocessing

## Show the output of the quality assesment on the TONO dataset

```python
import pandas as pd

df = pd.read_csv('./tono_assessment.csv', delimiter=';')
# Remove the ../../src/data/TONO/release/ prefix from the Filename column rows
df['Filename'] = df['Filename'].str.replace(r'../../src/data/TONO/release/', '', regex=True)
df['ContrastElement'] = df['Filename'].str.split('/').str[0]
df['Filename'] = df['Filename'].str.replace(r'^[^/]+/', '', regex=True)
df
```

```python
# Show the output of the ONOT assessment
df_onot = pd.read_csv('./onot_assessment.csv', delimiter=';')
# Remove the ../../src/data/TONO/release/ prefix from the Filename column rows
df_onot['Filename'] = df_onot['Filename'].str.replace(r'../../src/data/ONOT/S115379105/', '', regex=True)
df_onot['ContrastElement'] = None

df_onot
```

```python
# Merge the two dataframes
df_combined = pd.concat([df, df_onot], ignore_index=True)
# Move the ContrastElement as a second column
df_combined = df_combined[['Filename', 'ContrastElement'] + [col for col in df_combined.columns if col not in ['Filename', 'ContrastElement']]]
df_combined
```

```python
# Keep only the columns that output the scalar values, file names, and contrast elements
cols_to_keep = ['Filename', 'ContrastElement'] + [col for col in df_combined.columns if col.endswith('scalar')]
df_combined = df_combined[cols_to_keep]
df_combined
```

For each of the contrast element group, 6 images will be chosen with the variety of the utility score for providing the testing base for the LLM

```python
# In each of the contrast elements group, get 3 images with the highest UnifiedQualityScore.scalar, 3 with the lowest UnifiedQualityScore.scalar, and 3 with the median UnifiedQualityScore.scalar. or closest to it.
def get_top_bottom_median(group):
    top = group.nlargest(2, 'UnifiedQualityScore.scalar')
    bottom = group.nsmallest(2, 'UnifiedQualityScore.scalar')
    median_value = group['UnifiedQualityScore.scalar'].median()
    median = group.iloc[(group['UnifiedQualityScore.scalar'] - median_value).abs().argsort()[:2]]
    
    return pd.concat([top, bottom, median])

# Apply the function to each group of ContrastElement
df_in_context_dataset = df_combined.groupby('ContrastElement').apply(get_top_bottom_median).reset_index(drop=True)

# Add all the rows from the ONOT assesnment: where ContrastElement is None
df_in_context_dataset = pd.concat([df_in_context_dataset, df_combined[df_combined['ContrastElement'].isnull()]], ignore_index=True)

# Save the resulting DataFrame to a CSV file
df_in_context_dataset.to_csv('tono_assessment_contrast_element_test.csv', index=False)
```

# In-context learning

Provide the descirption to the chosen test data set

```python
# Read the saved CSV file
df_in_context_dataset = pd.read_csv('tono_assessment_contrast_element_test.csv')
df_in_context_dataset
```

Provide the descriptions based on the contrast element using manually added `description.py`

```python
from src.scripts.description import generate_description

def safe_generate_description(contrast_element):
    if pd.isna(contrast_element):
        return "The image is compliant for biometric use. No defects detected."
    return generate_description(contrast_element)

df_in_context_dataset['Description'] = df_in_context_dataset['ContrastElement'].apply(safe_generate_description)

show_columns = ['Filename', 'ContrastElement', 'UnifiedQualityScore.scalar', 'Description']

# Show the whole Description column output in the df using pandas
pd.set_option('display.max_colwidth', None)
df_in_context_dataset[show_columns]
```

```python
# Write the final DataFrame to a json file
df_in_context_dataset.to_json('incontext_test_data.json', orient='records', force_ascii=False, indent=4)
```
